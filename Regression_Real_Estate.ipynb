```python
area_drop=dataset.select_dtypes(include=['int64','float64'])
area_drop
```

```python
area_drop=area_drop.drop(['No'],axis = 1)
area_drop
```

```python
area_drop
x=area_drop.iloc[:,:-1].values
x
```

```python
y=area_drop.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
y_pred
```

```python
y_test
```

```python
df1 = pd.DataFrame({"actual price": y_test,"predicted price": y_pred})
df1
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
df1.plot(figsize=(20,10))
plt.show()
```

```python
plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
sns.scatterplot(x="actual price", y="predicted price", data = df1)

plt.subplot(1,2,2)
sns.regplot(x="actual price", y="predicted price", data = df1)
```
